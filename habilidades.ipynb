```python
import pandas as pd

contadores_habilidades2 = {
    'Python': 0,
    'R': 0,
    'SQL': 0,
    'NoSQL': 0,
    'MySQL': 0,
    'Oracle': 0,
    'Postgre': 0,
    'Power BI': 0,
    'Tableau': 0,
    'Scala': 0,
    'C++': 0,
    'Spark': 0,
    'office 365': 0,
    'SAS': 0,
    'Hadoop': 0,
    
    'GitHub':0,
    'Business Intelligence': 0,
    'Cloud': 0,
    
    'Java': 0,
    'Machine learning': 0,
    'Excel': 0,
    
    
    
    'Microsoft Azure': 0,
    'Amazon Web Services': 0,
    'Kafka': 0

   

}


for descripcion in df2.loc[:,'Job Description']:
    if pd.notna(descripcion):
        if ' bi ' in descripcion:
            contadores_habilidades2['Power BI'] += 1
        if 'python' in descripcion:
            contadores_habilidades2['Python'] += 1
        if ' r ' in descripcion:
            contadores_habilidades2['R'] += 1
        if 'sql ' in descripcion:
            contadores_habilidades2['SQL'] += 1
        if 'postgre' in descripcion:
            contadores_habilidades2['Postgre'] += 1
        
        
        if 'oracle' in descripcion:
            contadores_habilidades2['Oracle'] += 1
        if 'nosql' in descripcion:
            contadores_habilidades2['NoSQL'] += 1
        if 'mysql' in descripcion:
            contadores_habilidades2['MySQL'] += 1        
        if 'tableau' in descripcion:
            contadores_habilidades2['Tableau'] += 1
        if 'scala' in descripcion:
            contadores_habilidades2['Scala'] += 1
        if 'c++' in descripcion:
            contadores_habilidades2['C++'] += 1
        if 'spark' in descripcion:
            contadores_habilidades2['Spark'] += 1
        if 'office ' in descripcion:
            contadores_habilidades2['office 365'] += 1
        if ' sas ' in descripcion:
            contadores_habilidades2['SAS'] += 1
        if 'hadoop' in descripcion:
            contadores_habilidades2['Hadoop'] += 1
        if 'java' in descripcion:
            contadores_habilidades2['Java'] += 1
        if 'machine learning' in descripcion:
            contadores_habilidades2['Machine learning'] += 1
        if ' git' in descripcion:
            contadores_habilidades2['GitHub'] += 1    
        if 'business intelligence' in descripcion:
            contadores_habilidades2['Business Intelligence'] += 1
        if ' cloud ' in descripcion:
            contadores_habilidades2['Cloud'] += 1
        if 'excel ' in descripcion:
            contadores_habilidades2['Excel'] += 1
        
        if 'azure' in descripcion:
            contadores_habilidades2['Microsoft Azure'] += 1
        if 'amazon' in descripcion:
            contadores_habilidades2['Amazon Web Services'] += 1
        if 'kafka' in descripcion:
            contadores_habilidades2['Kafka'] += 1
        #if 'reinforcement' in descripcion:
            #contadores_habilidades2['Reinforcment'] += 1
        
        

for habilidad, cuenta in contadores_habilidades2.items():
    print(f"Número de trabajos que mencionan '{habilidad}': {cuenta}")
```

```python
df_skills = pd.DataFrame({'Habilidad Técnica': list(contadores_habilidades2.keys()), 'Demanda': list(contadores_habilidades2.values())})
len(df_skills)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df_habilidades = df_skills.sort_values(by='Demanda', ascending=False)

# Configurar el estilo de Seaborn
sns.set(style="whitegrid")

# Crear el gráfico de barras horizontales con tonalidad azul y valores
plt.figure(figsize=(10, 10))
sns.barplot(x='Demanda', y='Habilidad Técnica', data=df_habilidades)
sns.color_palette("mako", as_cmap=True)


plt.xlabel('Número de Ofertas')
plt.ylabel('Habilidad Técnica')
plt.title('Número de Ofertas para Habilidades Técnicas')

# Añadir los valores en las barras
for index, value in enumerate(df_habilidades['Demanda']):
    plt.text(value, index, str(value))

# Muestra el gráfico
plt.show()
```
