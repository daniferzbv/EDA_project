```python
import pandas as pd
from bs4 import BeautifulSoup

# Cargar el archivo CSV en un DataFrame
df = pd.read_csv('provinciacas.csv')

# Función para eliminar el texto HTML
def remove_html_tags(text):
    if isinstance(text, str):
        soup = BeautifulSoup(text, 'html.parser')
        return soup.get_text()
    return text

# Aplicar la función a todas las celdas del DataFrame
provinciacas = df.map(remove_html_tags)

# El DataFrame provinciaca ahora contiene los datos sin el texto HTML

provinciacas
```

```python
import pandas as pd

puesto_interes = 'Analista de datos / BI'

conteo_por_provincia = {}

for provincia in df.columns:
    
    ofertas_en_provincia = provinciacas[provincia]
    indice_puesto_interes = ofertas_en_provincia[ofertas_en_provincia == puesto_interes].index
   
    if not indice_puesto_interes.empty:
        
        indice = indice_puesto_interes[0]
        numero_ofertas = ofertas_en_provincia[indice + 1]
        conteo_por_provincia[provincia] = numero_ofertas

for provincia, conteo in conteo_por_provincia.items():
    print(f"{provincia}: {conteo}")


```

```python
df_conteo_analyst = pd.DataFrame(list(conteo_por_provincia.items()), columns=['Provincia', 'Ofertas'])

df_conteo_analyst
```
