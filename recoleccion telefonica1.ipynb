# SCRAPPING DE TELEFONICA 1

En este documento comienzo a scrappear la pagina del mapa de empleo de telefonica, consiguiendo datos de toda la peninsula en general y datos de trabajo en remoto

```python
from bs4 import BeautifulSoup as bs
import requests
import re
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup as bs
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import requests
```

SELENIUM

```python
driver = webdriver.Chrome()

url = "https://mapadelempleo.fundaciontelefonica.com/"

driver.get(url)
```

```python
cookies=  driver.find_element(By.XPATH, "/html/body/div[3]/div[2]/div/div/div[2]/div/div/button[2]")
cookies.click()
```

```python
Cargar_mas_profesiones = driver.find_element(By.XPATH, "/html/body/section[2]/div[1]/div/div/div/div[2]/div[1]/div[2]/div/button/p")
Cargar_mas_profesiones.click()
```

```python
# Espera un momento para asegurarte de que la página se haya cargado completamente
WebDriverWait(driver, 10).until(
    EC.presence_of_element_located((By.XPATH, "/html/body/section[2]/div[1]/div/div/div/div[2]/div[1]/div[2]/div/button/p"))
)

# Obtén el HTML de la página después de aceptar las cookies y cargar más profesiones
html = driver.page_source
```

BEAUTIFULSOUP

Pillamos las profesiones digitales y sus demandas a nivel nacional

```python
driver.quit()

# Ahora puedes utilizar BeautifulSoup para analizar el HTML
soup = BeautifulSoup(html, "lxml")

# Realiza el scraping en la página
profesiones_espana = soup.select('body section div.ranking-container p')
profesiones_espana
```

```python
espana_nombre=[]
espana_ofertas=[]
for dato in profesiones_espana:
    print(dato)
    espana_nombre.append(dato.get_text())
    espana_ofertas.append(dato.get_text())

espana_nombre=espana_nombre[0::2]
espana_ofertas=espana_ofertas[1::2]
```

```python
espana_ofertas
```

```python
dic_profesiones_espana={
'profesion': espana_nombre,
'puestos ofertados': espana_ofertas

    
}
```

```python
import pandas as pd
profesiones_espanapd=pd.DataFrame(dic_profesiones_espana)
profesiones_espanapd
```

```python
profesiones_espanapd.to_csv('profesiones_en_espana.csv', index=False)
```
