##### comparacion ofertas data VS resto

```python
import matplotlib.pyplot as plt


total_ofertas = df['puestos ofertados'].astype(int).sum()
profesiones_interes = ["Ingeniero/a de datos", "Analista de datos / BI", "Científico/a de datos"]
ofertas_data = df[df['profesion'].isin(profesiones_interes)]['puestos ofertados'].astype(int).sum()

porcentaje_data = (ofertas_data / total_ofertas) * 100


# Etiquetas para el gráfico
labels = ['', '',]

# Colores (paleta de azules)
colors = ['#1f77b4', '#aec7e8']  # Puedes personalizar los colores aquí

# Explotar la primera rebanada (resaltar)
explode = (0.1, 0)

# Cambiar el fondo a blanco
plt.figure(facecolor='white')

# Crea el gráfico
plt.pie([porcentaje_data, 100 - porcentaje_data], explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)

# Título
plt.title('Porcentaje de ofertas de profesiones data vs. otras ofertas', color='black')  # Texto en negro

# Muestra el gráfico
plt.axis('equal')  # Igualar la relación de aspecto para que el gráfico sea un círculo
plt.show()
```

```python
import pandas as pd
from bs4 import BeautifulSoup


df = pd.read_csv('provinciacas.csv')

# Función para eliminar el texto HTML
def remove_html_tags(text):
    if isinstance(text, str):
        soup = BeautifulSoup(text, 'html.parser')
        return soup.get_text()
    return text

# Aplicar la función a todas las celdas del DataFrame
provinciacas = df.map(remove_html_tags)

# El DataFrame provinciaca ahora contiene los datos sin el texto HTML

provinciacas
```

```python
import pandas as pd

puesto_interes = 'Analista de datos / BI'
conteo_por_provincia = {}

for provincia in df.columns:
   
    ofertas_en_provincia = provinciacas[provincia]
    indice_puesto_interes = ofertas_en_provincia[ofertas_en_provincia == puesto_interes].index
    
    if not indice_puesto_interes.empty:
        
        indice = indice_puesto_interes[0]
        numero_ofertas = ofertas_en_provincia[indice + 1]
        conteo_por_provincia[provincia] = numero_ofertas


for provincia, conteo in conteo_por_provincia.items():
    print(f"{provincia}: {conteo}")
```

```python
df_conteo_analyst = pd.DataFrame(list(conteo_por_provincia.items()), columns=['Provincia', 'Ofertas'])
df_conteo_analyst
```

```python
import plotly.express as px

# Datos con las provincias y los valores de interés
data = {
    'Provincia': [
        'madrid', 'barcelona', 'valencia', 'bizkaia', 'sevilla'
    ],
    'Ofertas': [605, 226, 66, 39, 29]
}

# Código ISO-ALPHA correspondiente para cada provincia
iso_alpha_mapping = {
    'madrid': 'ES-M',
    'barcelona': 'ES-B',
    'valencia': 'ES-V',
    'bizkaia': 'ES-BI',
    'sevilla': 'ES-SE'
}

# Mapear las provincias a los códigos ISO-ALPHA
data['ISO_ALPHA'] = [iso_alpha_mapping[provincia] for provincia in data['Provincia']]

# Crear el DataFrame
df = pd.DataFrame(data)

# Crear el gráfico con Plotly Express
fig = px.scatter_geo(df, locations="ISO_ALPHA", color="Provincia", size="Ofertas", locationmode="ISO-3", hover_name="Provincia")
fig.update_geos(projection_type="orthographic")
fig.update_layout(title="Número de Ofertas para Data Analyst por Provincia")
fig.show()
```
